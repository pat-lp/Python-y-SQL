### *GROUP BY, HAVING*

```python
import mysql.connector
```

```python
#Consulta utilizando group by+ having para determinar los clientes con el id de su dirección entre 10 y 50.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT customer_id, concat(first_name,' ',last_name), address_id FROM sakila.customer group by address_id having address_id between 10 and 50;")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON GROUP BY + HAVING")
    print("========================================================================")
    for fila in filas:
        print(f"ID CLIENTE: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]},\nID DIRECCIÓN: {fila[2]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Consulta group by+having de los clientes agrupados por montos y que los mismos sean mayor a 7.99 y menor 12.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT c.customer_id, concat(c.first_name,' ',c.last_name), p.amount FROM sakila.customer c inner join sakila.payment p on (c.customer_id = p.customer_id) group by p.amount having p.amount > 7.99 and p.amount < 12")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON GROUP BY + HAVING")
    print("========================================================================")
    for fila in filas:
        print(f"ID CLIENTE: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]},\nID DIRECCIÓN: {fila[2]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```
