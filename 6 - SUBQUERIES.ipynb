### *SUBCONSULTAS*

```python
import mysql.connector
```

```python
#Subconsulta para saber que clientes viven en Barcelona, Caracas y Dallas.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT customer_id, first_name, last_name FROM sakila.customer where address_id in(select address_id from sakila.address where city_id in(select city_id from sakila.city where city in ('Barcelona', 'Caracas','Dallas')))")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t SUBCONSULTAS")
    print("========================================================================")
    for fila in filas:
        print(f"ID CLIENTE: {fila[0]},\nNOMBRE: {fila[1]},\nAPELLIDO: {fila[2]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Subconsulta para saber que películas corresponden a la categoría: Acción, Comedia, Documental y Música.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT film_id, title FROM sakila.film where film_id in(select film_id from sakila.film_category where category_id in(select category_id from sakila.category where name in ('Action', 'Comedy','Documentary','Music')))")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t SUBCONSULTAS")
    print("========================================================================")
    for fila in filas:
        print(f"ID PELÍCULA: {fila[0]},\nTÍTULO PELÍCULA: {fila[1]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Subconsulta para saber que clientes alquilaron la película AGENT TRUMAN.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT customer_id, concat(first_name,' ',last_name) FROM sakila.customer where customer_id in(select customer_id from sakila.rental where inventory_id in(select inventory_id from sakila.inventory where film_id in(select film_id from sakila.film where title = 'AGENT TRUMAN')))")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t SUBCONSULTAS")
    print("========================================================================")
    for fila in filas:
        print(f"ID CLIENTE: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```
