### *SUM, MAX, MIN*

```python
import mysql.connector
```

```python
#Consulta utilizando inner join y min para determinar los clientes con el menor pago.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT c.customer_id, concat(c.first_name,' ',c.last_name), min(p.amount) as 'Suma mínima' FROM sakila.customer c inner join sakila.payment p on (c.customer_id = p.customer_id) group by customer_id")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON MIN")
    print("========================================================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]},\nMONTO MÍNIMO: {fila[2]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Consulta utilizando inner join y min para determinar los clientes con el mayor pago.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT c.customer_id, concat(c.first_name,' ',c.last_name), max(p.amount) as 'Suma mínima' FROM sakila.customer c inner join sakila.payment p on (c.customer_id = p.customer_id) group by customer_id")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON MAX")
    print("========================================================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]},\nMONTO MÍNIMO: {fila[2]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Consulta utilizando inner join y min para determinar los clientes con su total pagado.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT c.customer_id, concat(c.first_name,' ',c.last_name), round(sum(p.amount),1) as 'Suma mínima' FROM sakila.customer c inner join sakila.payment p on (c.customer_id = p.customer_id) group by customer_id order by sum(p.amount) desc")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON SUM")
    print("========================================================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]},\nMONTO TOTAL: {fila[2]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```
