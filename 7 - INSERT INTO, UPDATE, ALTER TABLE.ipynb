### *INSERT INTO, UPDATE, ALTER TABLE*

```python
import mysql.connector
```

```python
#Se agregan actores a la tabla sakila.actors.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()

    sql = ("insert into actor (first_name, last_name) values (%s,%s)")
    nuevos_actores = [('TOM','HANKS'),
                      ('PAUL','NEWMAN'),
                      ('MICHAEL','CAINE'),
                      ('DUSTIN', 'HOFFMAN'),
                      ('JACK', 'NICHOLSON')
                     ]
    
    cursor.executemany(sql, nuevos_actores)
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Se verifican los actores agregados.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("select actor_id, concat(first_name,' ',last_name) from actor order by actor_id desc limit 10")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("==============================================")
    print("\t\t CONSULTA SELECT")
    print("==============================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]}.")
        print("-------------------------------")
    print("==============================================")
    print("\t\t FIN CONSULTA")
    print("==============================================")

    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Se agrega la columna edad en la tabla sakila.actors.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
 
    cursor.execute("alter table actor add age int(11)")
  
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```
