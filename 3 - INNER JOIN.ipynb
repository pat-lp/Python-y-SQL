### *INNER JOIN*

```python
import mysql.connector
```

```python
#Consulta utilizando inner join para determinar la clasificación y el idioma de las películas, seleccionando sólo las categorias de películas 'Action', 'Music' y 'Sci-Fi'.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT f.film_id as ID, f.title as Título, f.description as Descripción, cat.name as Clasificación, l.name as 'Idioma' FROM sakila.film f inner join sakila.film_category c on (f.film_id = c.film_id) inner join sakila.category cat on (c.category_id = cat.category_id) inner join sakila.language l on (f.language_id = l.language_id) where cat.name in ('Action','Music','Sci-Fi')")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON INNER JOIN")
    print("========================================================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nTÍTULO: {fila[1]},\nDESCRIPCIÓN: {fila[2]},\nCLASIFICACIÓN: {fila[3]},\nIDIOMA: {fila[4]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Consulta de los clientes con las películas que seleccionaron.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT c.customer_id, c.first_name, c.last_name, f.title FROM sakila.customer c  inner join sakila.rental r on (c.customer_id = r.customer_id) inner join sakila.inventory i on (r.inventory_id = i.inventory_id) inner join sakila.film f on (i.film_id = f.film_id) order by c.customer_id")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON INNER JOIN")
    print("========================================================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nNOMBRE: {fila[1]},\nAPELLIDO: {fila[2]},\nPELICULA: {fila[3]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```

```python
#Consulta de los clientes con la cantidad de películas que seleccionaron.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("SELECT c.customer_id, concat(c.first_name,' ',c.last_name), count(f.title) FROM sakila.customer c  inner join sakila.rental r on (c.customer_id = r.customer_id) inner join sakila.inventory i on (r.inventory_id = i.inventory_id) inner join sakila.film f on (i.film_id = f.film_id) group by c.customer_id")
    
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("========================================================================")
    print("\t\t\t CONSULTA CON INNER JOIN")
    print("========================================================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nNOMBRE Y APELLIDO: {fila[1]},\nCANTIDAD PELÍCULAS ALQUILADAS: {fila[2]}.")
        print("-------------------------------")
        
    print("========================================================================")
    print("\t\t\t FIN CONSULTA")
    print("========================================================================")
    
    conexion.commit()
    
except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    conexion.close()
    print("\nCierre de conexión exitoso!!!")
```
