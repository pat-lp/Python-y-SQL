### *ORDER BY*

```python
import mysql.connector
```

```python
#Consulta utilizando alias y ordenando los actores por Apellido de forma ascendente.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    cursor.execute("select actor_id as id, first_name as Nombre, last_name as Apellido, last_update as Actualización from actor order by Apellido asc")
    filas = cursor.fetchall()

    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("==============================================")
    print("\t\t CONSULTA ORDER BY")
    print("==============================================")
    for fila in filas:
        print(f"ID: {fila[0]},\nNOMBRE: {fila[1]},\nAPELLIDO: {fila[2]}.")
        print("-------------------------------")
        
    print("==============================================")
    print("\t\t FIN CONSULTA")
    print("==============================================")

    conexion.commit()
    conexion.close()

except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    print("\nCierre de conexión exitoso!!!")
```

### *WHERE*


```python
#Consulta utilizando where para seleccionar actor por nombre y apellido cumpliendo con la condición del like.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    cursor.execute("select * from actor where first_name like '%N' and last_name like 'W%' ")
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("==============================================")
    print("\t\t CONSULTA WHERE Y AND")
    print("==============================================")
    for fila in filas:
        print(f"NOMBRE: {fila[1]},\nAPELLIDO: {fila[2]}.")
        print("-------------------------------")
    print("==============================================")
    print("\t\t FIN CONSULTA")
    print("==============================================")

    conexion.commit()
    conexion.close()

except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    print("\nCierre de conexión exitoso!!!")
    
```

```python
#Consulta utilizando where y la condición like+or para seleccionar actores por nombre y apellido.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()  
    cursor.execute("select * from actor where first_name like '%N' or last_name like 'W%' ")  
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("==============================================")
    print("\t\t CONSULTA WHERE Y OR")
    print("==============================================")
    for fila in filas:
        print(f"NOMBRE: {fila[1]},\nAPELLIDO: {fila[2]}.")
        print("-------------------------------")
    print("==============================================")
    print("\t\t FIN CONSULTA")
    print("==============================================")

    conexion.commit()
    conexion.close()

except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    print("\nCierre de conexión exitoso!!!")
```

```python
#Consulta utilizando where y like+not para la selección de actores por nombre y apellido.
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()  
    cursor.execute("select * from actor where first_name like '%N' and last_name not like 'W%' ")
    filas = cursor.fetchall()
    
    print(f"\n\nRESULTADOS OBTENIDOS\nTotal de registros: {cursor.rowcount}\n")
    print("==============================================")
    print("\t\t CONSULTA WHERE Y NOT")
    print("==============================================")
    for fila in filas:
        print(f"NOMBRE: {fila[1]},\nAPELLIDO: {fila[2]}.")
        print("-------------------------------")
    print("==============================================")
    print("\t\t FIN CONSULTA")
    print("==============================================")

    conexion.commit()
    conexion.close()

except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
finally:
    print("\nCierre de conexión exitoso!!!")
```
