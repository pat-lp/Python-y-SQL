### *ORDER BY*

```python
import mysql.connector
```

```python
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    cursor.execute("select actor_id as id, first_name as Nombre, last_name as Apellido, last_update as ActualizaciÃ³n from actor order by Apellido asc")
    filas = cursor.fetchall()

    print("==============================================")
    print("\t\t CONSULTA ORDER BY")
    print("==============================================")
    for fila in filas:
        id, nombre, apellido, act = fila
        print(f"ID: {id}, NOMBRE: {nombre}, APELLIDO: {apellido}")
        
    print("==============================================")
    print("\t\t FIN CONSULTA")
    print("==============================================")

    conexion.commit()
    conexion.close()

except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
```

### *WHERE*


```python
try:
    conexion = mysql.connector.connect(host='localhost', user='root',password='',database='sakila')
    cursor = conexion.cursor()
    
    cursor.execute("select * from actor where first_name like '%N' and last_name like 'W%' ")
    
    filas = cursor.fetchall()
    
    
    print("==============================================")
    print("\t\t CONSULTA WHERE Y AND")
    print("==============================================")
    for fila in filas:
        id, nombre, apellido, act = fila
        print(f"NOMBRE: {nombre}, APELLIDO: {apellido}")
        
    print("==============================================")
    print("\t\t FIN CONSULTA")
    print("==============================================")

    conexion.commit()
    conexion.close()

except Exception as e:
    print(f"ERROR DE CONEXION--> {e}")
    
    
```
